```python
# Import necessary modules
import numpy as np
import matplotlib.pyplot as plt

# Declare a list
my_list = [1, 2, 3, 4, 5]

# Declare a NumPy array
my_array = np.array([1, 2, 3, 4, 5])

# Define a basic Python function
def square(x):
  return x ** 2

# Apply the function to the list and array
squared_list = [square(x) for x in my_list]
squared_array = square(my_array)

# Create a simple Matplotlib figure
plt.figure()
plt.plot(my_list, squared_list, label='List Squared')
plt.plot(my_array, squared_array, label='Array Squared')
plt.xlabel('Original Values')
plt.ylabel('Squared Values')
plt.title('Square of List and Array Values')
plt.legend()
plt.savefig('squared_values.png')
plt.show()
```
